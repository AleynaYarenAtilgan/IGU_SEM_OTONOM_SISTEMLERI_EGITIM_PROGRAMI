```python
#List veri tipi ornegi-1/ Bir listenin elemanlarinin toplam,maksimum ve minimum degerleri
sayilar=[10,20,30,40,50]

#Toplama hesaplama
toplam= sum(sayilar)
print("Toplam:",toplam)

#Maksimum ve minimum deger bulma
print("En buyuk sayi:",max(sayilar))
print("En kucuk sayi:",min(sayilar))
```
