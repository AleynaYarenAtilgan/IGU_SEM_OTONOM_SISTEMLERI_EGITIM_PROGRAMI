```python
ogrenciler=(
    ("Ali",(85,90,78)),
    ("Ayşe",(88,76,92)),
    ("Mehmet",(70,85,80)),
    ("Zeynep",(95,90,93)),
)

#öğrenci not hesaplama
for ogrenci in ogrenciler:
  isim=ogrenci[0]
  notlar=ogrenci[1]
  ortalama=sum(notlar)/len(notlar)
  print(isim,"adlı öğrencinin ortalaması",ortalama)
```
